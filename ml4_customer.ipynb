```python
gender = pd.get_dummies(df['Gender'], drop_first=True)
gender
```

```python
geo = pd.get_dummies(df['Geography'], drop_first=True)
geo
```

```python
df = pd.concat([df,gender,geo],axis=1)
df.drop(['Gender','Geography','RowNumber','CustomerId','Surname'],inplace=True,axis=1)
df.head()
```

```python
df.shape
```

```python
X = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(activation="relu",input_dim=11,units=6,kernel_initializer="uniform"))
classifier.add(Dense(activation="relu",units=6,kernel_initializer="uniform"))
classifier.add(Dense(activation="sigmoid",units=1,kernel_initializer="uniform"))
```

```python
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
classifier.summary()
```

```python
classifier.fit(X_train,y_train,batch_size=10,epoch=50)
```

```python
F_pred = classifier.predict(X_test)
F_pred
```

```python
y_pred = (F_pred>0.5)
y_pred
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```
